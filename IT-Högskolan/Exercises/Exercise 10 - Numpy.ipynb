## 1. Dice Simulations (*)

 - a) ' Calculate theoretical mean of a six sided die. (*) '
 
The sum of each side in a die is 21. 
The mean should approach 21 / 6 ~ 3.5.

```python
from random import randint
import matplotlib.pyplot as plt
import numpy as np

def DiceThrower(Throws): # I've modified a similar function from earlier exercises.
    DiceArray = []
    for Throw in range(Throws):
        DiceArray.append(randint(1,6)) 
    return DiceArray # Return a list, since np.mean wants a list of numbers.


DiceSumDictionary = {}
ThrowList = list(map(int, (np.logspace(1,7,7))))  
# np.logspace(1,7,7) ~= [1.0e7, 1.0e6, etc..] (numpy.float64) -> map int to all 'floats' and return map object.
# Then turn map object back into a list. It's now [10, 100, 1000, 10000, 100000, 1000000, 10000000].

for Throws in ThrowList:
    DiceSumDictionary[Throws] = np.mean(DiceThrower(Throws))

plt.title("Resulting means from a six-sided dice.")    
plt.xlabel("Throws")
plt.ylabel("Mean")
plt.plot(DiceSumDictionary.keys(), DiceSumDictionary.values(), marker='o')

xAxisValues = list(DiceSumDictionary.keys())
xAxisRange = list(range(len(xAxisValues)))
print(xAxisValues)
print(xAxisRange)

plt.xticks(xAxisRange, xAxisValues)

print(plt.xticks())

```
